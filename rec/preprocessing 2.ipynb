# Identify Genre Types

```python
g_types = ['fiction', 'nonfiction', 'horror', 'thriller', 'fantasy', 'mystery', 'history', 'romance', 'classic', 'crime', 'comedy']

def isNonfiction(x):
    labels = ["nonfiction"]
    if (any(label in x for label in labels)):
        return True
    return False
def isFiction(x):
    labels = ["fiction"]
    if (any(label in x for label in labels)):
        return True
    return False
def isHorror(x):
    labels = ["horror"]
    if (any(label in x for label in labels)):
        return True
    return False
def isThriller(x):
    labels = ["thriller"]
    if (any(label in x for label in labels)):
        return True
    return False
def isFantasy(x):
    labels = ["fantasy"]
    if (any(label in x for label in labels)):
        return True
    return False
def isMystery(x):
    labels = ["mystery"]
    if (any(label in x for label in labels)):
        return True
    return False
    return False
def isHistory(x):
    labels = ["history", "historical"]
    if (any(label in x for label in labels)):
        return True
    return False
def isRomance(x):
    labels = ["romance", "romantic", "love"]
    if (any(label in x for label in labels)):
        return True
    return False
def isClassic(x):
    labels = ["classic"]
    if (any(label in x for label in labels)):
        return True
    return False
def isCrime(x):
    labels = ["crime"]
    if (any(label in x for label in labels)):
        return True
    return False
def isComedy(x):
    labels = ["humour", "funny", "comedy"]
    if (any(label in x for label in labels)):
        return True
    return False
```

# Split genres with group encoding

```python
ins = pd.DataFrame(columns=g_types)
df1 = pd.concat([df1, ins])

print(df1.columns)

df1.to_csv('rg.csv', encoding='utf-8', index=False)
```

```python
d = pd.read_csv("rg.csv")

for idx, entry in enumerate(d.genres):
    string = d.genres[idx]
#     print(f"idx: {idx}    string: {string}")
    
    if (isNonfiction(string)):
        d.loc[idx,"nonfiction"] = 1
    elif (isFiction(string)):
        d.loc[idx,"fiction"] = 1
    if (isHorror(string)):
        d.loc[idx,"horror"] = 1
    if (isThriller(string)):
        d.loc[idx,"thriller"] = 1
    if (isFantasy(string)):
        d.loc[idx,"fantasy"] = 1
    if (isMystery(string)):
        d.loc[idx,"mystery"] = 1
    if (isHistory(string)):
        d.loc[idx,"history"] = 1
    if (isRomance(string)):
        d.loc[idx,"romance"] = 1
    if (isClassic(string)):
        d.loc[idx,"classic"] = 1
    if (isCrime(string)):
          d.loc[idx,"crime"] = 1
    if (isComedy(string)):
        d.loc[idx,"comedy"] = 1

    
display(d)
```
