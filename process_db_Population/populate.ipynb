```python
import pandas as pd
import numpy as np

df = pd.read_csv("final.csv")
```

```python
df.columns
df.shape
```

```python
 genres = {
    "mystery": 0,
    "fiction": 1,
    "nonfiction": 2,
    "thriller": 3,
    "romance": 4,
    "horror": 5,
    "history": 6,
    "classic" : 7,
    "crime": 8,
    "fantasy": 9,
    "comedy": 10
}

book_cols = ['bookID', 'isbn', 'title', 'description', 'publisher', 'published', 'coverPath', 'ratings', 'avgRating', 'readers']
genre_cols = ['bookID', 'isbn', 'genre']


old_cols = ['isbn', 'title', 'desc', 'publisher', 'publication', 'rating_count', 'rating']
```

```python
# g0 = df[old_g]
# display(g0)

new = df[old_cols]
display(new)
```

```python
new.insert(5, 'coverPath', None)
new.insert(8, 'readers', 0)
display(new)
```

```python
new['publication'] = new['publication'].astype('int64')
new['rating_count'] = new['rating_count'].astype('int64')
new = new[new['publication'] > 0]
display(new)

# make copy of df for genre and author
df2 = df.copy()

df2['publication'] = df2['publication'].astype('int64')
df2['rating_count'] = df2['rating_count'].astype('int64')
df2 = df2[df2['publication'] > 0]
display(df2)
```

```python
new['length'] = df.desc.str.len()
```

```python
# truncate the description column values
new.insert(2, 'desc2', '')

for i, row in new.iterrows():
    if row['length'] >= 495:
        x = row['desc'][:495] + "..."
        new.at[i,'desc2'] = x
    else:
        new.at[i,'desc2'] = row['desc']
```

```python
# display(new)
# for i, row in new.iterrows():
#     print(len(row.desc2))

print(new.columns)

cols = ['isbn', 'title', 'desc2', 'publisher', 'publication', 'coverPath', 'rating_count', 'rating', 'readers']

# create new dataframe with desired values for book table
new1 = new[cols]
print(new1.columns)
```

```python
# Initialise bookID starting from 3 for both dataframes
# bookID to be used as PK for books, bookAuthors and bookGenres

df2.insert(0, 'bookID', range(3, 3 + len(df2)))
new1.insert(0, 'bookID', range(3, 3 + len(new1)))
print(new1.shape)
print(df2.shape)
```

```python
# rename columns for book table
new1.columns = book_cols
print(new1.columns)
# change datetime format
new1.published = pd.to_datetime(new1.published, format='%Y')
display(new1)
# save to csv
new1.to_csv("book_populate.csv",index=False)
```

```python
n = 1000
count = 1
for g, df in new1.groupby(np.arange(len(new1)) // n):
    name = "books_" + str(count) + ".csv"
    df.to_csv(name,index=False)
    print(df.shape)
    count += 1
```

```python
rows_list = []
# for row in input_rows:

#         dict1 = {}
#         # get input row in dictionary format
#         # key = col_name
#         dict1.update(blah..) 

#         rows_list.append(dict1)

# df = pd.DataFrame(rows_list) 


# for every row
for i, row in g0.iterrows():
    # check if row has any positive genre entry
    for genre in genres.keys():
        # if contains positive entry, add dict to list of rows for bookGenres dataframe
        if row[genre] == 1:
            g_dict = dict()
            g_dict['bookID'] = row['bookID']
            g_dict['genre'] = genres[genre]
            rows_list.append(g_dict)

g1 = pd.DataFrame(rows_list)
```

```python
display(g1)
g1.bookID = g1.bookID.astype('int64')
g1.genre = g1.genre.astype('int64')
display(g1)
```

```python
# save to csv
g1.to_csv("genres_populate.csv",index=False)

# Set authors csv
n = 3000
count = 1
for g, df in g1.groupby(np.arange(len(g1)) // n):
    name = "genres_" + str(count) + ".csv"
    df.to_csv(name,index=False)
    print(df.shape)
    count += 1
```
